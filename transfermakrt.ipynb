```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
name=[]
position=[]
age=[]
nation=[]
team=[]
value=[]

matches=[]
goals=[]
owngoals=[]
assists=[]
yellowcard=[]
secondyc=[]
redcard=[]
subin=[]
subout=[]
```

```python
df = pd.DataFrame(
        {"name":name,
         "position":position,
         "age":age,
         "nation":nation,
         "team":team,
         "value":value,
         "matches": matches,
         "goals":goals,
         "owngoals":owngoals,
         "assists":assists,
         "yellowcard":yellowcard,
         "second yellow card":secondyc,
         "redcard":redcard,
         "substitutions in":subin,
         "substitutions out":subout}
    )
df
```

```python
df['value'] = df['value'].str.replace('€', '')
df['value'] = df['value'].str.replace('m', '')
df['value'] = df['value'].str.replace('k', '000')
```

```python
df['value'] = df['value'].str.replace('.*k', '', regex = True)
```

```python
df
```

```python
df['position'] = df['position'].str.replace('Goalkeeper', 'GK')
df['position'] = df['position'].str.replace('Centre-Back', 'DF')
df['position'] = df['position'].str.replace('Right-Back', 'DF')
df['position'] = df['position'].str.replace('Left-Back', 'DF')
df['position'] = df['position'].str.replace('Defender', 'DF')
df['position'] = df['position'].str.replace('Attacking Midfield', 'MF')
df['position'] = df['position'].str.replace('Defensive Midfield', 'MF')
df['position'] = df['position'].str.replace('Left Midfield', 'MF')
df['position'] = df['position'].str.replace('Right Midfield', 'MF')
df['position'] = df['position'].str.replace('midfield', 'MF')
df['position'] = df['position'].str.replace('Central Midfield', 'MF')
df['position'] = df['position'].str.replace('Left Winger', 'FW')
df['position'] = df['position'].str.replace('Centre-Forward', 'FW')
df['position'] = df['position'].str.replace('Second Striker', 'FW')
df['position'] = df['position'].str.replace('Attack', 'FW')
df['position'] = df['position'].str.replace('Right Winger', 'FW')
df['position'].unique()
```

```python
df.to_csv('trans.csv', index=False) # 저장 후 다시 불러오기
```

```python
# 몸값을 억 단위로 변경
df = pd.read_csv('trans.csv')
df.loc[df['value'] >= 1000, 'value'] = df['value'] / 1000000
df['value'] = df['value'] * 15
df
```

```python
df['name'] = df['name'].str.lower()
df
```

```python
# 한국 선수들 이름 형식 변경
df.loc[df['name'].str.contains('.*-.* .*') & df['nation'].str.contains('Korea, South'), 'name'] = df["name"].apply(lambda x: " ".join(reversed(x.split(" "))))
```

```python
df.to_csv('trans.csv', index=False)
```

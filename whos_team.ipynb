```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import time
import pandas as pd
import numpy as np
```

```python
teamname=[]
goals=[]
SPG=[]
PS=[]
aerialswon=[]
rating=[]
blocks=[]
tackles=[]
interceptions=[]
fouls=[]
offside=[]
SOTPG=[]
Drb=[]
fouled=[]
```

```python
url = 'https://1xbet.whoscored.com/Regions/81/Tournaments/6/Seasons/9121/Stages/21027/TeamStatistics/Germany-2-Bundesliga-2022-2023'
chrome_driver = 'C:\Rselenium\chromedriver'
driver = webdriver.Chrome(chrome_driver)
driver.get(url)
driver.implicitly_wait(5)
```

```python
driver.find_element(By.XPATH, '/html/body/div[4]/div[6]/div[1]/ul/li[2]/a').click()
time.sleep(2)
driver.find_element(By.XPATH, '/html/body/div[4]/div[6]/div[1]/ul/li[3]/a').click()
time.sleep(2)
driver.find_element(By.XPATH, '/html/body/div[4]/div[6]/div[1]/ul/li[1]/a').click()
time.sleep(2)

table1 = driver.find_element(By.XPATH, '/html/body/div[4]/div[6]/div[2]/div[3]')
table2 = driver.find_element(By.XPATH, '/html/body/div[4]/div[6]/div[4]/div[3]')
table3 = driver.find_element(By.XPATH, '/html/body/div[4]/div[6]/div[3]/div[3]')

table1_html= table1.get_attribute('innerHTML')
table2_html= table2.get_attribute('innerHTML')
table3_html= table3.get_attribute('innerHTML')

soup1 = BeautifulSoup(table1_html, 'html.parser')
soup2 = BeautifulSoup(table2_html, 'html.parser')
soup3 = BeautifulSoup(table3_html, 'html.parser')

player_info1= soup1.find_all('tr')
player_info2= soup2.find_all('tr')
player_info3= soup3.find_all('tr')
```

```python
for i in range(1, len(player_info1)):
    player_sum = player_info1[i].find_all("td")
    player_def = player_info2[i].find_all("td")
    player_off = player_info3[i].find_all("td")
    teamname.append(player_sum[0].text)
    goals.append(player_sum[1].text)
    SPG.append(player_sum[2].text)
    PS.append(player_sum[5].text)
    aerialswon.append(player_sum[6].text)
    rating.append(player_sum[7].text)
    blocks.append(player_def[1].text)
    tackles.append(player_def[2].text)
    interceptions.append(player_def[3].text)
    fouls.append(player_def[4].text)
    offside.append(player_def[5].text)
    SOTPG.append(player_off[2].text)
    Drb.append(player_off[3].text)
    fouled.append(player_off[4].text)
```

```python
df = pd.DataFrame(
        {"teamname":teamname,
         "goal":goals,
         "SPG":SPG,
         "PS":PS,
         "aerialswon":aerialswon,
         "blocks":blocks,
         "tackles":tackles,
         "intecept":interceptions,
         "foul":fouls,
         "offside":offside,
         "SOTPG":SOTPG,
         "Drb":Drb,
         "fouled":fouled,
         "rating":rating}
    )
df
```

```python
df['teamname'] = df['teamname'].str.replace('.*\. ', '', regex=True)
df
```

```python
df.to_csv('whoscored_team.csv', header=False, index=False, mode = 'a')
```
